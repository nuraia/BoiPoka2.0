```python
#########creating matrix##########

dataset = dataset.drop_duplicates(['User', 'Book_id'])
user_book_matrix = dataset.pivot(index='User', columns='Book_id', values='Rating')
user_book_matrix.fillna(0, inplace=True)

user_book_matrix.head(10)

```

Singular Value Decomposition (SVD) to break the user-book matrix into user vector and book vector (matrix factorization)


```python
import numpy as np
import pandas as pd
from scipy.sparse.linalg import svds
from scipy.sparse import csc_matrix

def build_SVD_embeddings(user_book_matrix, embedding_length):
    # Convert user_book_matrix to a sparse matrix (csc format)
    user_book_matrix_sparse = csc_matrix(user_book_matrix)

    # Build SVD embeddings to be used later on
    U, Sigma, VT = svds(user_book_matrix_sparse, k=embedding_length)
    user_embed_df = pd.DataFrame(U, index=user_book_matrix.index)
    VT_T = np.transpose(VT)
    book_embed_df = pd.DataFrame(VT_T, index=user_book_matrix.columns)

    return user_embed_df, book_embed_df
```

```python
user_book_matrix = user_book_matrix.astype(float)
```

```python
embedding_length = 64
user_embed_df, book_embed_df = build_SVD_embeddings(user_book_matrix, embedding_length)
```

Get full vector by combining user traits and book traits

```python
def append_embeddings(df, user_embed_df, book_embed_df):
    #Append embeddings to the full list of rated user-book combinations
    #Returns df with appended embeddings [user] + [book] in a column
    def find_user_book(x):
        #Functions used to
        user_row = user_embed_df.loc[user_embed_df.index == x[0]]
        book_row = book_embed_df.loc[book_embed_df.index == x[1]]
        return np.concatenate((np.array(user_row),np.array(book_row)), axis = None)

    df['Full-Vector'] = df[['User','Book_id']].apply(find_user_book, axis = 1)

    return df
```

```python
dataset= append_embeddings(dataset, user_embed_df, book_embed_df)

dataset.head()


```

Splitting the data for train, test. Our input feature is the full vector, target is book rating

```python


def create_train_test(df, y_col = 'Rating'): #target variable
    #Manipulates the df with Full-Vector to the appropriate format needed for NN modeling.
    #X_train, X_test, y_train, y_test data ready to be fed into NN model
    X_total = df['Full-Vector'] #input feature

    X_total = np.stack((X_total))

    y_total = np.array(df[y_col])
    y_total = np.expand_dims(y_total,axis = 1)

    from sklearn.model_selection import train_test_split

    X_train, X_test, y_train, y_test = train_test_split(X_total, y_total, test_size = 0.2 , random_state = 21)

    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = create_train_test(dataset)
```

```python
#NN Model


import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Input, BatchNormalization, Dropout, Lambda

def baseline_model(total_embed_length):
    total = Input((total_embed_length))


    d1 = Dense(2048, activation = 'relu')(total)
    d1 = Dropout(0.2)(d1)
    d1 = BatchNormalization()(d1)

    d2 = Dense(1024, activation = 'relu')(d1)
    d2 = Dropout(0.2)(d2)
    d2 = BatchNormalization()(d2)

    d3 = Dense(512, activation = 'relu')(d2)
    d3 = Dropout(0.2)(d3)
    d3 = BatchNormalization()(d3)

    d4 = Dense(256, activation = 'relu')(d3)
    d5 = Dense(64, activation = 'relu')(d4)

    d6 = Dense(16, activation = 'relu')(d5)

    out = Dense(1, activation = 'relu')(d6)

    model = Model(inputs=[total], outputs=[out])
    #compile model outside of this function to make it flexible.
    model.summary()
    return model


```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, concatenate

# Assuming embedding_length is defined
embedding_length = 64

# Train the baseline model
baseline_input = Input(shape=(embedding_length*2,))
baseline_model = baseline_model(embedding_length*2)
baseline_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error'])
baseline_model.summary()
baseline_model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

```

```python
import pandas as pd


def create_book_title_list(df):
    # Create a DataFrame with unique ISBNs and their titles
    book_ids = pd.DataFrame(df[['Book_id', 'Book_Name']].drop_duplicates())
    return book_ids

# Create a DataFrame containing unique ISBNs and their corresponding titles
book_IDS = create_book_title_list(dataset)

# Display the first 10 rows of the resulting DataFrame
print(book_IDS.head(10))
```

```python
def get_top_matches(user_embed_df, book_embed_df, combined_final, book_merged, query_index, model, k_matches=5):
    #storing the input userID, storing it in sampled_user_ID
    sampled_user_vector = user_embed_df.iloc[[query_index]]
    sampled_user_ID = sampled_user_vector.index[0]
    #Reassigns sampled_user_vector as a NumPy array
    sampled_user_vector = np.array(user_embed_df.iloc[query_index])
     #Initializes an empty DataFrame called prediction_df with the index from book_embed_df.
    prediction_df = pd.DataFrame(index=book_embed_df.index)
    prediction_df.reset_index(inplace=True)
    #dads a 'User' column to prediction_df with the sampled user ID.
    prediction_df['User'] = sampled_user_ID


#contructs a full vector for each book and sample user vector

    def build_prediction_embeddings(x):
        book_row = book_embed_df.loc[book_embed_df.index == x]
        return np.concatenate((sampled_user_vector, np.array(book_row)), axis=None)


#Applies the build_prediction_embeddings function to each book ID in prediction_df and stores in the 'Full-Vector' column.
    prediction_df['Full-Vector'] = prediction_df['Book_id'].apply(build_prediction_embeddings)
#the full vector in prediction dataframe is my input
    X_to_train = np.stack(prediction_df['Full-Vector'])
 #train the model and store it in Y
    Y_to_train = model.predict(X_to_train)
#Adds a 'Trained_Items' column to prediction_df containing the predicted ratings.
    prediction_df['Trained_Items'] = Y_to_train
#Filters the DataFrame to include only the rows where the user has already rated books.
    already_read = combined_final[combined_final['User'] == sampled_user_ID]
#Selects rows in prediction_df corresponding to books that the user has already rated.
    top_5_unrated = prediction_df.loc[prediction_df['Book_id'].isin(already_read['Book_id'])]
#Sorts the top_5_unrated DataFrame based on the 'Trained_Items' column in descending order. (not trained)
    top_5_unrated.sort_values(by=['Trained_Items'], inplace=True, ascending=False)
    top_5_unrated = top_5_unrated.iloc[0:k_matches]
    # Include Book_Name in top_unrated
    top_5_unrated['Book_Name'] = top_5_unrated['Book_id'].apply(lambda x: book_merged.loc[book_merged['Book_id'] == x, 'Book_Name'].values[0])




    return top_5_unrated
```

```python
# Sample index for the user you want to get top matches for
sampled_user_index = 34  # Change this to the desired user index

# Call the function to get top matches using the first trained model
top_unrated = get_top_matches(
    user_embed_df, book_embed_df, combined_final, book_merged, sampled_user_index, baseline_model, k_matches=5
)



print("\nTop Unrated Books:")
print(top_unrated)
```
